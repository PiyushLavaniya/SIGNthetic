#### Importing the Dependencies

```python
import os
import time
import cv2
import shutil
import numpy as np
import json
import tqdm
from IPython.display import FileLink
```

### Data Preparatation
#### Data contains the following informaiton:
1. Split
2. frame_end
3. frame_start
4. video_path
5. gloss

```python
##Reading the JSON File
with open("Data\WLASL_v0.3.json", "r") as json_file:
    all_data = json.load(json_file)
```

```python
##Length of the Data
len(all_data)
```

#### Extracting Important information from the Data

```python
videos_directory = "Data/videos"

##Data Structure to store all the data
data = []

for i in tqdm.tqdm(range(len(all_data)), ncols=100):
    
    ##Taking out gloss and instances from the all_data
    
    gloss = all_data[i]["gloss"]
    instances = all_data[i]["instances"]
    
    ##Now we loop over all the instances and extract the required information
    for instance in instances:
        video_id = instance["video_id"]
        
        if os.path.exists(os.path.join(videos_directory, f"{video_id}.mp4")):
            video_path = os.path.join(videos_directory, f"{video_id}.mp4")
            
        else:
            continue
        
        frame_start = instance["frame_start"]
        frame_end = instance["frame_end"]
        split = instance["split"]
    
        ##Appending this information to the data Data Structure
        data.append({
            "frame_start": frame_start,
            "frame_end": frame_end,
            "split": split,
            "video_path": video_path,
            "gloss": gloss
        })
```

#### Saving the Organized Data for further use-cases and training

```python
##Saving the JSON file since the Data is an object
with open("Reorganized_Data.json", "w") as json_file:
    json.dump(data, json_file, indent = 4)
```
